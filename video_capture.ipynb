```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture("video_1.mp4")
fgbg1 = cv2.createBackgroundSubtractorMOG2()
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('Output1.avi', fourcc, 30, (640, 480))
x1,y1 = 0,0
x2,y2 = 0,0
a, b = 0,0
c, d = 0,0
```

```python
while True:
    
    ret, frame = cap.read()
    h, w = frame.shape[0:2]
    #print("height", h)
    #print("width", w)
    black_image = np.zeros((h,w,3), np.uint8)
    #cv2.imshow("Frame", frame)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    frame = fgbg1.apply(frame)
    cv2.imshow("Gray", gray)
    gray = cv2.GaussianBlur(gray, (3,3), 0)
    #cv2.imshow("Gray", gray)
    ret1, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    
    def centroid(contours, image):
        global y1,y2,a,b,x1,x2,c,d
        if cv2.contourArea(contours)>1000:
            cent_moment = cv2.moments(contours)
            centroid_x = int(cent_moment['m10']/cent_moment['m00'])
            centroid_y = int(cent_moment['m01']/cent_moment['m00'])
            if centroid_y<500 or centroid_x>220:
                cv2.circle(image, (centroid_x,centroid_y), 4,(255,0,0), -1)      
                cv2.putText(image, '(' + str(centroid_x) + ',' + str(centroid_y) + ')' , (centroid_x-105,centroid_y-5), cv2.FONT_HERSHEY_SIMPLEX, 1, (centroid_x-110,centroid_y))
            if centroid_x<420:
                y1 = centroid_y
                x1 = centroid_x
                Vy = int((y1-a)/0.02)
                Vx = int((x1-c)/0.02)
                if Vy>0 and (x1>210 or y1<500):
                    cv2.line(image, (x1, y1), (x1, y1+120), (255,255,255),1)
                    cv2.line(image, (x1, y1+120), (x1+5, y1+115), (0,0,255),2)
                    cv2.line(image, (x1, y1+120), (x1-5, y1+115), (0,0,255),2)
                    
                    cv2.putText(image, "Vy = " + str(Vy) , (x1, y1+120), cv2.FONT_HERSHEY_SIMPLEX, 1, (x1+10, y1+130))
                elif Vy<0 and (x1>210 or y1<500):
                    cv2.line(image, (x1, y1), (x1, y1-120), (255,255,255),1)
                    cv2.line(image, (x1, y1-120), (x1+5, y1-115), (0,0,255),2)
                    cv2.line(image, (x1, y1-120), (x1-5, y1-115), (0,0,255),2)
                    
                    cv2.putText(image, "Vy = " + str(Vy) , (x1, y1-120), cv2.FONT_HERSHEY_SIMPLEX, 1, (x1, y1-90))
                elif Vy == 0:
                    pass
                a = y1
                if Vx>0 and (x1>210 or y1<500):
                    cv2.line(image, (x1, y1), (x1+100, y1), (255,255,255),1)
                    cv2.line(image, (x1+100, y1), (x1+95, y1-5), (255,255,0),2)
                    cv2.line(image, (x1+100, y1), (x1+95, y1+5), (255,255,0),2)
                    cv2.putText(image, "Vx = " + str(Vx) , (x1+100, y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (x1+100, y1+10))
                    
                elif Vx<0 and (x1>210 or y1<500):
                    cv2.line(image, (x1, y1), (x1-140, y1), (255,255,255),1)
                    cv2.line(image, (x1-140, y1), (x1-135, y1-5), (255,255,0),2)
                    cv2.line(image, (x1-140, y1), (x1-135, y1+5), (255,255,0),2)
                    cv2.putText(image, "Vx = " + str(Vx) , (x1-250, y1-30), cv2.FONT_HERSHEY_SIMPLEX, 1, (x1-230, y1-50))
                elif Vx == 0:
                    pass
                c = x1
            if centroid_x>420 :
                y2 = centroid_y
                x2 = centroid_x
                Uy = int((y2-b)/0.02)
                Ux = int((x2-d)/0.02)
                if Uy>0 and y2<500:
                    cv2.line(image, (x2, y2), (x2, y2+120), (255,255,255),1)
                    cv2.line(image, (x2, y2+120), (x2+5, y2+115), (0,0,255),2)
                    cv2.line(image, (x2, y2+120), (x2-5, y2+115), (0,0,255),2)
                    cv2.putText(image, "Uy = " + str(Uy) , (x2, y2+120), cv2.FONT_HERSHEY_SIMPLEX, 1, (x2+10, y2+130))  
                    
                elif Uy<0 and y2<500:
                    cv2.line(image, (x2, y2), (x2, y2-120), (255,255,255),1)
                    cv2.line(image, (x2, y2-120), (x2+5, y2-115), (255,255,0),2)
                    cv2.line(image, (x2, y2-120), (x2-5, y2-115), (255,255,0),2)
                    cv2.putText(image, "Uy = " + str(Uy) , (x2, y2-120), cv2.FONT_HERSHEY_SIMPLEX, 1, (x2+10, y2-130))
                elif Uy == 0:
                    pass
                b = y2
                if Ux>0:
                    cv2.line(image, (x2, y2), (x2+120, y2), (255,255,255),1) 
                    cv2.line(image, (x2+120, y2), (x2+115, y2-5), (255,255,0),2)
                    cv2.line(image, (x2+120, y2), (x2+115, y2+5), (255,255,0),2)
                    cv2.putText(image, "Ux = " + str(Ux) , (x2+120, y2), cv2.FONT_HERSHEY_SIMPLEX, 1, (x2+130, y2-10))
                elif Ux<0:
                    cv2.line(image, (x2, y2), (x2-120, y2), (255,255,255),1)
                    cv2.line(image, (x2-120, y2), (x2-115, y2-5), (255,255,0),2)
                    cv2.line(image, (x2-120, y2), (x2-115, y2+5), (255,255,0),2)
                    cv2.putText(image, "Ux = " + str(Ux) , (x2-120, y2), cv2.FONT_HERSHEY_SIMPLEX, 1, (x2-130, y2-10))
                elif Ux == 0:
                    pass
                d = x2
            
    
              
    def contour(contours, image):
        if cv2.contourArea(contours)>1000:
            cv2.drawContours(image, contours, -1, (0,255,0), -2)
        else:
            pass
    edg = cv2.Canny(threshold, 10, 100)
    #cv2.imshow("Trial", edg)
    dup = edg.copy()
    contours, hierarchy = cv2.findContours(dup, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    #print(str(len(contours)))            
    for i in contours:
        contour(i, black_image)
        centroid(i, black_image)
        #print("j=", str(j)) 
    if ret == True:    
        cv2.imshow("Video", black_image)
        out.write(black_image)
    key = cv2.waitKey(30)
    if key ==27:
        break 
cap.release()
out.release()
cv2.destroyAllWindows()
                                 
                    
                    
                    
                    
                    
                    
                    
                    
```
